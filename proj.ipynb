```r
library(tidyverse)
library(repr)
library(tidymodels)
source("cleanup.R")
```

```r
player_data <- read_csv("players.csv")
player_data
```

```r
tidy_player_data <- player_data |>
select(played_hours, subscribe, experience)

player_summary <- tidy_player_data |>
group_by(experience) |>
summarize(subscribe_rate = mean(subscribe == TRUE))

player_plot <- ggplot(tidy_player_data, aes(x = experience, y = played_hours, colour = subscribe)) +
  geom_point(alpha = 0.6) +
  labs(title = "Subscribed Players Across Experience Levels and Hours Played", x = "Experience", y = "Hours Played", colour = "Subscribed")

player_bar <- ggplot(player_summary, aes(x = experience, y = subscribe_rate, fill = experience)) +
geom_bar(stat = "identity") +
labs(title = "Subscription Rate by Experience", x = "Experience", y = "Proportion Subscribed") +
scale_y_continuous(labels = scales::percent)

player_plot
player_bar
```

```r
mean_hours <- player_data |>
summarize(avg_played_hours = mean(played_hours))

mean_age <- player_data |>
summarize(avg_age = mean(Age))

player_table <- tibble(mean_hours, mean_age)
player_table
```
